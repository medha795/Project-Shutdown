```python
#Loading the csv file into pandas DataFrame 
import pandas as pd 
df = pd.read_csv("RBLSF.csv") 
df.head()
```

```python
# Get column names
column_names = df.columns
print(column_names)
```

```python
#drop columns that you recognize as unnecessary 
df.drop(["Location_Id","Business_Account_Number","Ownership_Name","Mail_Address","Mail_State", "Mail_City","Mail_Zipcode","Parking_Tax","Transient_Occupancy_Tax","LIC_Code","LIC_Code_Description","Supervisor_District","Business_Corridor","UniqueID","SF_Find_Neighborhoods","Current_Police_Districts","Current_Supervisor_Districts","Analysis_Neighborhoods","Neighborhoods"], axis = 1, inplace = True) 
  
df
```

```python
#Get details of the data that is not based in San Francisco 
indexNames = df[ df['City'] != ("San Francisco" or "San+Francisco")].index 
# Delete these row indexes from dataFrame
df.drop(indexNames,inplace=True)
df
```

```python
#extract the DataFrame to an output csv file 
df.to_csv('extracted_data.csv', sep='\t', encoding='utf-8')
```
